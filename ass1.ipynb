# Assignment 1 (Traffic dataset)

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading Data:

```python
data = pd.read_csv("traffic.csv")
```

```python
data
```

### question 1:

```python
fig, axis = plt.subplots(2, 2, figsize = (12, 8))

axis[0, 0].boxplot(data["CarCount"])
axis[0, 0].set_title("CarCount")

axis[0, 1].boxplot(data["BikeCount"])
axis[0, 1].set_title("BikeCount")

axis[1, 0].boxplot(data["BusCount"])
axis[1, 0].set_title("BusCount")

axis[1, 1].boxplot(data["TruckCount"])
axis[1, 1].set_title("TruckCount")

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (14, 10))

sns.histplot(data = data, x = "CarCount", ax = ax[0, 0]).set(title = "CarCount", xlabel=None, ylabel=None)
sns.histplot(data = data, x = "BikeCount", ax = ax[0, 1]).set(title = "BikeCount", xlabel=None, ylabel=None)
sns.histplot(data = data, x = "BusCount", ax = ax[1, 0]).set(title = "BusCount", xlabel=None, ylabel=None)
sns.histplot(data = data, x = "TruckCount", ax = ax[1, 1]).set(title = "TruckCount", xlabel=None, ylabel=None)

plt.show()
```

### question 2:

```python
labels = ['normal', 'heavy', 'high', 'low']
plt.pie(list(data["Traffic Situation"].value_counts()), labels = labels, autopct='%.0f%%')
plt.title("Traffic Situation counts")
plt.show()
```

Mostly the traffic situation being normal

### question 3:

```python
fig, ax = plt.subplots(2, 2, figsize = (14, 14))

sns.barplot(x = 'Day of the week', y =  "CarCount", ax = ax[0, 0], data = data).set(title = "CarCount", xlabel=None, ylabel=None)

sns.barplot(x = 'Day of the week', y =  "BikeCount", ax = ax[0, 1], data = data).set(title = "BikeCount", xlabel=None, ylabel=None)

sns.barplot(x = 'Day of the week', y =  "BusCount", ax = ax[1, 0], data = data).set(title = "BusCount", xlabel=None, ylabel=None)

sns.barplot(x = 'Day of the week', y =  "TruckCount", ax = ax[1, 1], data = data).set(title = "TruckCount", xlabel=None, ylabel=None)

for axis in ax.flat:
    axis.tick_params(axis='x', rotation=45)

plt.show()
```

The car counts nearly equals in all days, but the othe vehicles not, the Bike Count reach the peek on friday while Bus and truck being less.

### question 14:

```python
vehicle_types = ['CarCount', 'BikeCount', 'BusCount', 'TruckCount']
variances = {vehicle: data[vehicle].var() for vehicle in vehicle_types}

plt.bar(variances.keys(), variances.values(), color=['blue', 'orange', 'green', 'red'])
plt.title('Variance in Vehicle Counts Across Vehicle Types')
plt.xlabel('Vehicle Type')
plt.ylabel('Variance')
plt.show()
```

Cars is the most vehicle type used

### question 15:

```python
# Group the data by 'Time' and calculate the mean for each vehicle type
average_vehicle_count = data.groupby('Time')[['CarCount', 'BikeCount', 'BusCount', 'TruckCount']].mean()

# Plot the average vehicle count over time
plt.figure(figsize=(14, 8))
average_vehicle_count.plot(kind='line', figsize=(14, 8))
plt.title('Average Vehicle Count Over Time')
plt.xlabel('Time')
plt.ylabel('Average Count')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Type')
plt.show()
```

The number of cars increase between 5:30 AM to 8:30 AM
