# Assignment 1 (Traffic dataset)

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading Data:

```python
data = pd.read_csv("traffic.csv")
```

```python
data
```

### question 1:

```python
fig, axis = plt.subplots(2, 2, figsize = (12, 8))

axis[0, 0].boxplot(data["CarCount"])
axis[0, 0].set_title("CarCount")

axis[0, 1].boxplot(data["BikeCount"])
axis[0, 1].set_title("BikeCount")

axis[1, 0].boxplot(data["BusCount"])
axis[1, 0].set_title("BusCount")

axis[1, 1].boxplot(data["TruckCount"])
axis[1, 1].set_title("TruckCount")

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize = (14, 10))

sns.histplot(data = data, x = "CarCount", ax = ax[0, 0]).set(title = "CarCount", xlabel=None, ylabel=None)
sns.histplot(data = data, x = "BikeCount", ax = ax[0, 1]).set(title = "BikeCount", xlabel=None, ylabel=None)
sns.histplot(data = data, x = "BusCount", ax = ax[1, 0]).set(title = "BusCount", xlabel=None, ylabel=None)
sns.histplot(data = data, x = "TruckCount", ax = ax[1, 1]).set(title = "TruckCount", xlabel=None, ylabel=None)

plt.show()
```

### question 2:

```python
labels = ['normal', 'heavy', 'high', 'low']
plt.pie(list(data["Traffic Situation"].value_counts()), labels = labels, autopct='%.0f%%')
plt.title("Traffic Situation counts")
plt.show()
```

### question 3:

```python
sns.histplot(x = 'Day of the week', data = data)
plt.xlabel('Day')
plt.ylabel('Total vehicles count')
plt.title('Toatal vehicles in each day of week')
plt.xticks(rotation=45)
plt.show()
```

### questions (4, 5, 6, 7):

```python
fig, ax = plt.subplots(2, 2, figsize = (14, 10))

sns.violinplot(x = 'Traffic Situation', y = 'CarCount', data = data, ax = ax[0, 0]).set(title = "car count and traffic situation", xlabel=None, ylabel=None)

sns.violinplot(x = 'Traffic Situation', y = 'BikeCount', data = data, ax = ax[0, 1]).set(title = "bike count and traffic situation", xlabel=None, ylabel=None)

sns.violinplot(x = 'Traffic Situation', y = 'BusCount', data = data, ax = ax[1, 0]).set(title = "bus count and traffic situation", xlabel=None, ylabel=None)

sns.violinplot(x = 'Traffic Situation', y = 'TruckCount', data = data, ax = ax[1, 1]).set(title = "truck count and traffic situation", xlabel=None, ylabel=None)

plt.show()
```

### question 8:

```python
sns.histplot(x = 'Traffic Situation', data = data)
plt.title('Total vehicles count with Traffic Situation')
plt.ylabel("Total vehicles count")
plt.xticks(rotation=45)
plt.show()
```

### question 9:

```python
df = pd.DataFrame(data[['Time', 'Total']].groupby('Time').sum()).sort_values('Total', ascending=False)[:10]
```

```python
df
```

```python
sns.barplot(x = 'Time', y = 'Total', data = df)
plt.title("Top 10 Busiest Hours of the Day")
plt.xlabel('Time')
plt.ylabel("Total vehicles")
plt.xticks(rotation=45)
plt.show()
```
